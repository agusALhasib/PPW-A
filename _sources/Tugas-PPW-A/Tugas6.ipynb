# Label Encoder
Mengonversi data kategorikal (teks) menjadi angka menggunakan **Label Encoding**. Berikut penjelasan langkah-langkahnya:

1. **Inisialisasi LabelEncoder**:
   - `label_encoder = LabelEncoder()` membuat objek `LabelEncoder` untuk mengubah data teks menjadi representasi numerik.

2. **Proses Transformasi**:
   - `label_encoder.fit_transform(df['kategori'])`:
     - Mengambil kolom `kategori` dalam DataFrame `df`.
     - Mengubah setiap kategori unik (seperti "olahraga", "makanan") menjadi angka yang sesuai (seperti 0, 1, 2, dll.).
   - Hasil encoding disimpan ke kolom baru bernama `kategori_encoded` di DataFrame `df`.

3. **Menampilkan Nilai Sebelum dan Sesudah Konversi**:
   - `label_encoder.classes_`: Menampilkan semua kategori unik di kolom `kategori`.
   - `label_encoder.transform(label_encoder.classes_)`: Menampilkan angka hasil encoding untuk setiap kategori.
   - Hasilnya ditampilkan sebagai pasangan key-value (kategori asli -> angka).

4. **Menampilkan DataFrame Hasil Encoding**:
   - `df.head()` menampilkan lima baris pertama dari DataFrame `df`, dengan kolom tambahan `kategori_encoded` yang berisi hasil encoding.



```python
from sklearn.preprocessing import LabelEncoder

# Transformasi data kategorik
label_encoder = LabelEncoder()
df.loc[:, 'kategori_encoded'] = label_encoder.fit_transform(df['kategori'])

# Menampilkan nilai sebelum dan sesudah konversi
print("\nNilai sebelum dan sesudah konversi:")
print(dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_))))

print("\nDataFrame setelah Label Encoding:")
df.head()
```

Berikut pustaka Python untuk memproses data teks, membangun model klasifikasi, dan mengevaluasi kinerjanya:

1. **`pandas`**: Untuk membaca dan mengelola data dalam format tabel (DataFrame).
2. **`train_test_split`**: Membagi data menjadi data latih (untuk melatih model) dan data uji (untuk menguji model).
3. **`TfidfVectorizer`**: Mengubah teks menjadi angka dengan teknik TF-IDF (mengukur pentingnya kata dalam dokumen).
4. **`TruncatedSVD`**: Mengurangi jumlah fitur dari data teks agar lebih sederhana.
5. **`LogisticRegression`**: Model klasifikasi yang memprediksi kategori berdasarkan data.
6. **`classification_report` & `accuracy_score`**: Mengevaluasi performa model dengan metrik seperti akurasi dan F1-score.

Kode ini adalah bagian dari proses pembelajaran mesin untuk analisis teks dan klasifikasi.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

 memisahkan **fitur** dan **label** dari dataset:

1. **`x`**: Berisi data teks yang sudah melalui proses _stopword removal_ (penghapusan kata-kata umum yang tidak memiliki makna penting), digunakan sebagai **fitur/input** untuk model.
2. **`y`**: Berisi label numerik hasil encoding dari kolom **`kategori`**, digunakan sebagai **output/target** yang ingin diprediksi oleh model.

Dengan pemisahan ini, data teks (`x`) akan diproses lebih lanjut untuk melatih model agar dapat memprediksi label (`y`).

```python
# Memisahkan fitur dan label
x = df['stopword_removal']
y = df['kategori_encoded']
```

Membagi dataset menjadi dua bagian: **data latih** dan **data uji**, dengan rincian:

1. **`x_train`**: Fitur (teks) untuk melatih model.
2. **`x_test`**: Fitur (teks) untuk menguji kinerja model.
3. **`y_train`**: Label target untuk melatih model.
4. **`y_test`**: Label target untuk mengukur akurasi prediksi pada data uji.

**Parameter:**
- **`test_size=0.2`**: 20% dari data digunakan sebagai data uji, sedangkan 80% untuk data latih.
- **`random_state=42`**: Menjamin pembagian data yang konsisten (reproducible) setiap kali kode dijalankan.

**Tujuan**: Memastikan model dilatih pada sebagian data (latih) dan diuji pada data terpisah (uji) untuk mengevaluasi performanya.

```python
# Membagi data menjadi data latih dan data uji
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

**representasi teks ke dalam vektor TF-IDF** dan mengonversinya menjadi bentuk DataFrame agar lebih mudah dipahami.

### Penjelasan:
1. **TF-IDF (Term Frequency-Inverse Document Frequency)**:
   - Mengubah teks menjadi representasi numerik berdasarkan seberapa sering suatu kata muncul dalam dokumen dibandingkan dengan seluruh koleksi dokumen.

2. **Langkah-langkah:**
   - **`fit_transform(x_train)`**: Menghitung nilai TF-IDF untuk teks dalam **data latih** dan mengubahnya menjadi matriks.
   - **`transform(x_test)`**: Menggunakan model TF-IDF yang sama untuk menerapkan transformasi pada **data uji**.
   - **`get_feature_names_out()`**: Mendapatkan daftar kata (fitur) yang digunakan untuk representasi TF-IDF.

3. **Konversi ke DataFrame:**
   - Mengubah hasil TF-IDF (yang berbentuk matriks sparse) menjadi DataFrame untuk **data latih (`df_train_tfidf`)** dan **data uji (`df_test_tfidf`)**.
   - Kolom DataFrame merepresentasikan kata-kata unik dari teks.

4. **Output:**
   - Menampilkan **DataFrame TF-IDF untuk data latih**, dengan baris merepresentasikan dokumen dan kolom menunjukkan nilai TF-IDF untuk setiap kata.


### Tujuan:
Mengubah teks menjadi bentuk numerik untuk digunakan dalam tahap selanjutnya, seperti klasifikasi atau analisis.

```python
# Tahap 1: TF-IDF
tfidf_vectorizer = TfidfVectorizer()
x_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
x_test_tfidf = tfidf_vectorizer.transform(x_test)

# Mengonversi hasil TF-IDF menjadi DataFrame
df_train_tfidf = pd.DataFrame(x_train_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
df_test_tfidf = pd.DataFrame(x_test_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
print("Hasil TF-IDF (Data Latih):")
df_train_tfidf
```

*menggunakan metode SVD (Singular Value Decomposition)** untuk mereduksi dimensi data yang dihasilkan dari proses TF-IDF, lalu mengubah hasilnya menjadi DataFrame agar lebih mudah dianalisis.

### Penjelasan:
1. **Reduksi Dimensi dengan SVD**:
   - **TruncatedSVD** adalah algoritma yang digunakan untuk mereduksi dimensi data matriks TF-IDF.
   - Tujuannya adalah mempertahankan informasi penting dari data dalam jumlah dimensi yang lebih kecil, sehingga analisis lebih efisien.

2. **Parameter Penting**:
   - **`n_components=100`**: Mengatur jumlah dimensi baru menjadi 100 (dari ribuan kata sebelumnya).
   - **`random_state=42`**: Menjamin hasil pengacakan konsisten untuk eksperimen yang dapat direproduksi.

3. **Proses:**
   - **`fit_transform(x_train_tfidf)`**: Menyesuaikan model SVD dengan data latih TF-IDF dan mengubahnya menjadi representasi berdimensi lebih kecil.
   - **`transform(x_test_tfidf)`**: Menerapkan model SVD yang sama pada data uji TF-IDF.

4. **Konversi ke DataFrame**:
   - **`df_train_svd`**: DataFrame hasil reduksi dimensi untuk data latih.
   - **`df_test_svd`**: DataFrame hasil reduksi dimensi untuk data uji.
   - Setiap kolom dalam DataFrame adalah fitur baru yang telah diringkas, dan setiap baris adalah representasi dokumen.

5. **Output:**
   - Menampilkan hasil SVD untuk data latih dalam bentuk DataFrame, dengan 100 kolom (fitur baru yang diringkas).


```python
# Tahap 2: SVD (Reduksi Dimensi)
svd = TruncatedSVD(n_components=100, random_state=42)
x_train_svd = svd.fit_transform(x_train_tfidf)
x_test_svd = svd.transform(x_test_tfidf)

# Mengonversi hasil SVD menjadi DataFrame
df_train_svd = pd.DataFrame(x_train_svd)
df_test_svd = pd.DataFrame(x_test_svd)
print("\nHasil SVD (Data Latih):")
df_train_svd
```

**menggunakan Logistic Regression** sebagai model untuk mengklasifikasikan data setelah proses reduksi dimensi dengan SVD. Berikut penjelasan singkatnya:

### Penjelasan:
1. **Membuat Model Logistic Regression**:
   - **`LogisticRegression(C=0.2, random_state=42)`**:
     - **`C=0.2`**: Mengatur regularisasi untuk menghindari overfitting. Nilai kecil seperti 0.2 menambah penalti pada model agar tidak terlalu kompleks.
     - **`random_state=42`**: Menjamin hasil konsisten saat pelatihan model untuk eksperimen yang dapat direproduksi.

2. **Melatih Model dengan Data Latih**:
   - **`clf.fit(x_train_svd, y_train)`**:
     - Menggunakan fitur hasil reduksi dimensi dari data latih (`x_train_svd`).
     - Melibatkan label target (`y_train`) untuk membangun model klasifikasi.

3. **Prediksi pada Data Uji**:
   - **`y_pred = clf.predict(x_test_svd)`**:
     - Model yang telah dilatih digunakan untuk memprediksi kelas target dari data uji (`x_test_svd`).
     - Hasil prediksi disimpan di variabel **`y_pred`**.

### Tujuan:
- **Membuat klasifikasi otomatis** berdasarkan fitur dokumen.
- Model Logistic Regression bertindak sebagai **pengklasifikasi**, menggunakan data latih untuk mempelajari hubungan antara fitur dokumen (dari SVD) dan label kategori.


```python
# Tahap 3: Klasifikasi dengan Logistic Regression
clf = LogisticRegression(C=0.2, random_state=42)
clf.fit(x_train_svd, y_train)
y_pred = clf.predict(x_test_svd)
```

```python
# Evaluasi Model
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# Pengujian
Kode ini membuat dan menyimpan **pipeline** untuk proses machine learning yang terdiri dari tiga langkah:

1. **TF-IDF Vectorization**: Mengubah teks menjadi representasi numerik.
2. **SVD (Reduksi Dimensi)**: Mengurangi jumlah fitur dari data teks.
3. **Logistic Regression**: Melakukan klasifikasi untuk memprediksi kategori teks.

Setelah pipeline dilatih dengan data latih, pipeline disimpan dalam file `.pkl` menggunakan `joblib`. Ini memungkinkan penggunaan kembali pipeline yang sudah dilatih untuk prediksi tanpa perlu pelatihan ulang.

---



```python
import joblib
from sklearn.pipeline import Pipeline

# Misalnya, Anda telah membuat pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('x_train_svd', TruncatedSVD(n_components=100)),
    ('clf', LogisticRegression())
])

# Setelah melatih model
pipeline.fit(x_train, y_train)

# Menyimpan pipeline
joblib.dump(pipeline, 'svd_tfidf_pipeline.pkl')
```
